```python
# So by the structure of the file, each lab is splitted by '\t', and the phones will be the 3rd position splitted by ':'

## Commented to better visualization of notebook
# with open(lab_path[0], 'r', encoding='utf-8') as f: 
#     for line in f.readlines():
#         print(line)

# Printing normalized text and phones
with open(lab_path[0], 'r', encoding='utf-8') as f: 
    # print words
    lines = f.readlines()
    print(lines[1].split(':'))
    print(lines[2].split(':'))
```

```python
# Lets save phonetic transcription and normalized text that is the 2nd position splitted by ':'

normalized_texts = []
phonetic_transcription = []

wav_fixed = []
lab_fixed = []
speakers_fixed = []

# Pattern text to appear in the first position of words and phones transcriptions
word_pattern = '# words'
phone_pattern = '# phones'

k = 0
for lab in lab_path:
    try:
        with open(lab, 'rt', encoding='utf-8') as f:
        
            lines = f.readlines()
            
            # Index of where word and phone are
            word_idx = None
            phone_idx = None
            
            # For each line in the lab, search for words tag, if not, word_check = 'none'
            for i in range(len(lines)): 
                if('words' in lines[i].split(':')[0]):
                    word_check = lines[i].split(':')[0]
                    word_idx = i
                    break
                else:
                    word_check = 'none'

            # For each line in the lab, search for phones tag, if not, phone_check = 'none'
            for j in range(len(lines)):
                if('phones' in lines[j].split(':')[0]):
                    phone_check = lines[j].split(':')[0]
                    phone_idx = j
                    break
                else:
                    phone_check = 'none'

            if(word_check == word_pattern and phone_check == phone_pattern): 
                # [1:-1] means starting by the second position and ending in the N-1 position, because [0] = ' ' and [N] = '\n'
                normalized_texts.append(lines[word_idx].split(':')[1][1:-1])
                phonetic_transcription.append(lines[phone_idx].split(':')[1][1:-1])
                lab_fixed.append(lab_path[k])
                wav_fixed.append(wav_path[k])
                speakers_fixed.append(speakers[k])
                k += 1
            else:
                print(word_check, phone_check)
                print(lines)
                print('There were no word and phone transcription the same as the patterns. \n')
#                 break
                
    except:
        print(f"In lab file: {lab} \nUTF-8 was not able to decode, trying with iso-8859-1")
        
        with open(lab, 'rt', encoding='iso-8859-1') as f:
        
            lines = f.readlines()

            # Index of where word and phone are
            word_idx = None
            phone_idx = None
            
            # For each line in the lab, search for words tag, if not, word_check = 'none'
            for i in range(len(lines)): 
                if('words' in lines[i].split(':')[0]):
                    word_check = lines[i].split(':')[0]
                    word_idx = i
                    break
                else:
                    word_check = 'none'

            # For each line in the lab, search for phones tag, if not, phone_check = 'none'
            for j in range(len(lines)):
                if('phones' in lines[j].split(':')[0]):
                    phone_check = lines[j].split(':')[0]
                    phone_idx = j
                    break
                else:
                    phone_check = 'none'

            if(word_check == word_pattern and phone_check == phone_pattern): 
                # [1:-1] means starting by the second position and ending in the N-1 position, because [0] = ' ' and [N] = '\n'
                normalized_texts.append(lines[word_idx].split(':')[1][1:-1])
                phonetic_transcription.append(lines[phone_idx].split(':')[1][1:-1])
                lab_fixed.append(lab_path[k])
                wav_fixed.append(wav_path[k])
                speakers_fixed.append(speakers[k])
                k += 1
            else:
                print(word_check, phone_check)
                print(lines)
                print('There were no word and phone transcription the same as the patterns. \n')
#                 break
```

```python
# Lets just guarantee that the [0] and [n-1] position are "#" in phonetic transcriptions
def check_transcriptions(trans):
    if(trans[0] == "#" and trans[len(trans)-2] == '#'): # -2 means that python starts list with 0 so is N-1 the length, -1 is the position we want
        return True
    return False
```
